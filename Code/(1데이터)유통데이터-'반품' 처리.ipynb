## 0. 1데이터 시트 합치기(기간 합치기)

```python
import pandas as pd
```

```python
file_path = '(1 데이터) 유통데이터 활용 경진대회 배포용.xlsx'

# 모든 시트 합치기
excel_file = pd.ExcelFile(file_path)
df = pd.concat([pd.read_excel(file_path, sheet_name=sheet) for sheet in excel_file.sheet_names])

df
```

## 1. '반품' 행 처리 - 상품바코드별 계산 

```python
# '상품 바코드(대한상의)'로 그룹화하고 '판매수량'을 더함
df1 = df.groupby(['상품 바코드(대한상의)'], as_index=False).agg({
    '판매수량': 'sum',
    '판매일': 'first',   
    '구분': 'first',
    '우편번호': 'first',
    '매출처코드': 'first',
    '옵션코드': 'first',
    '규격': 'first',
    '입수': 'first',
    '상품명': 'first',
    '대분류': 'first',
    '중분류': 'first',
    '소분류': 'first'
})

df1
```

## 2. '반품' 행 처리 - 판매일, 상품바코드별 계산 

```python
# '판매일'과 '상품 바코드(대한상의)'로 그룹화하고 '판매수량' 더하기
df2 = df.groupby(['판매일', '상품 바코드(대한상의)'], as_index=False).agg({
    '판매수량': 'sum',      
    '구분': 'first',
    '우편번호': 'first',
    '매출처코드': 'first',
    '옵션코드': 'first',
    '규격': 'first',
    '입수': 'first',
    '상품명': 'first',
    '대분류': 'first',
    '중분류': 'first',
    '소분류': 'first'
})

df2
```

## 3. '반품' 행 처리 - 단순 반품 행 제거

```python
# '구분' 칼럼이 '반품'인 행을 제거
df3 = df[df['구분'] != '반품']

df3
```
