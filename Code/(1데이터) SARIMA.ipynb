```python
# 판매일 컬럼 제외
df_use_cr = df_use_add.drop(columns=['판매일'])

# 상관계수 계산
association = df_use_cr.corr()['판매수량']

# 절댓값 기준 내림차순 정렬
association_sorted = association.abs().sort_values(ascending=False)

# ASSOCIATION 값 출력
association_sorted
```

```python
# 0.1 이상인 변수 선택
selected_features = association_sorted[association_sorted >= 0.1].index.tolist()

# df_use에서 선택된 변수들만 남기기
df_filtered = df_use_add[selected_features + ['판매일']]  # 판매일도 포함하여 원본 데이터 유지

# 컬럼 순서 변경: '판매일', '판매수량'이 첫번째와 두번째 컬럼이 되도록
df_filtered = df_filtered[['판매일', '판매수량'] + selected_features]

# df_filtered에서 첫 번째 '판매수량' 컬럼 삭제
df_filtered = df_filtered.loc[:, ~df_filtered.columns.duplicated()]

# 결과 출력
df_filtered
```

```python
df_p = df_filtered
```

```python
# '판매일'을 datetime 형식으로 변환
df_p.loc[:, '판매일'] = pd.to_datetime(df_p['판매일'])

# 월 단위 집계: 판매수량은 합계, 다른 컬럼은 평균
df_p = df_p.groupby(pd.Grouper(key='판매일', freq='W')).agg(
    {'판매수량': 'sum',  # 판매수량 합계
     **{col: 'mean' for col in df_p.columns if col != '판매일' and col != '판매수량'}}  # 다른 컬럼 평균
).reset_index()


df_p
```

```python
import pandas as pd
from pmdarima import auto_arima
from sklearn.metrics import mean_squared_error
import numpy as np

# 데이터프레임을 불러온 후 '판매일' 컬럼을 datetime 형식으로 변환
df_p['판매일'] = pd.to_datetime(df_p['판매일'])

# 인덱스를 '판매일'로 설정
df_p.set_index('판매일', inplace=True)

# 월별로 집계 (판매수량의 합계)
df_p = df_p.resample('W').sum()

# '판매수량' 컬럼을 'sales'로 변경
df_p.rename(columns={'판매수량': 'sales'}, inplace=True)

data = df_p  # 모든 컬럼 포함

# 훈련 데이터와 테스트 데이터로 분할
train = data.iloc[:-26]
test = data.iloc[-26:]

# 'sales' 컬럼 제외한 설명변수 선택
exogenous_train = train.drop(columns=['sales'])
exogenous_test = test.drop(columns=['sales'])

# 최적의 SARIMA 모델 찾기 (exogenous 변수 추가)
sarima_model = auto_arima(
    train['sales'],
    exogenous=exogenous_train,
    seasonal=True,
    m=12,  # 월 단위 계절 주기
    start_p=0, start_q=0,
    max_p=5, max_q=5,
    start_P=0, start_Q=0,
    max_P=5, max_Q=5,  # 계절 AR과 MA의 최대 차수 증가
    d=None, D=None,  # 비계절 차분과 계절 차분을 자동으로 결정
    seasonal_test='ocsb',  # 계절성 검정
    trace=True,
    error_action='ignore',
    suppress_warnings=True,
    stepwise=True,
    n_jobs=1
)

# 예측 수행
forecast, conf_int = sarima_model.predict(n_periods=26, exogenous=exogenous_test, return_conf_int=True)

# 성능 평가
mse = mean_squared_error(test['sales'], forecast)
print(f'Mean Squared Error: {mse}')

# 예측 결과 및 신뢰구간 출력
forecast_df = pd.DataFrame({'Forecast': forecast, 'Lower CI': conf_int[:, 0], 'Upper CI': conf_int[:, 1]}, index=test.index)
print(forecast_df)
```

```python
from sklearn.metrics import mean_absolute_error

# 최적의 SARIMA 모델로 예측
sarima_forecast = sarima_model.predict(n_periods=26, exogenous=exogenous_test.values)  # exogenous 변수 사용

# RMSE 계산
sarima_rmse = np.sqrt(mean_squared_error(test['sales'], sarima_forecast))

# MAE 계산
sarima_mae = mean_absolute_error(test['sales'], sarima_forecast)

# 결과 출력
print(f'SARIMA 모델: {sarima_model.order}x{sarima_model.seasonal_order}, RMSE: {sarima_rmse}, MAE: {sarima_mae}')
```

```python
# SARIMA 예측 결과를 DataFrame으로 변환
sarima_forecast_df = pd.DataFrame(sarima_forecast, columns=['forecast'])
sarima_forecast_df['ds'] = pd.date_range(start='2024-01-01', periods=len(sarima_forecast), freq='W')

# 월 단위로 집계
sarima_forecast_df['month'] = sarima_forecast_df['ds'].dt.to_period('M')
monthly_sum = sarima_forecast_df.groupby('month')['forecast'].sum().reset_index()

# 결과 출력
print("Monthly Sum of SARIMA Forecast from January to June 2024:")
print(monthly_sum)
```
